```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
def generate_process(dt = 0.1, theta = 1.2, mu = 0.5, sigma = 0.3, n = 10000):
    x = np.zeros(n)
    for t in range(1,n):
        x[t] = x[t-1] + theta * (mu - x[t-1])*dt + sigma * normal(0, np.sqrt(dt)) #using np.sqrt(dt) because numpy takes std deviation
    return x

def plot_process(x):
    plt.plot(x)
    plt.xlabel('t')
    plt.ylabel('x[t]')
    plt.title('Ornstein Ulenbeck Process')
    plt.show()
```

```python
plot_process(generate_process())
```
