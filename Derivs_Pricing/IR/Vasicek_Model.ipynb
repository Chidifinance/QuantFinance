```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
```

```python
def vasicek_model(r0, kappa, theta, sigma, T=1, N = 1000):
    dt = T/float(N) #converting T in years to fractal steps dt based on n size
    t = np.linspace(0, T, N+1) #creates timestamps up to T, with size N+1
    rates = [r0] #List data structure starting at r0

    for _ in range(N):
        # dr = kappa * (theta - rates[-1]) * dt + sigma * normal(0, np.sqrt(dt)) #try np.sqrt(dt) * normal()
        dr = kappa * (theta - rates[-1]) * dt + sigma * np.sqrt(dt) * normal()
        rates.append(rates[-1] + dr)
    return t, rates

def plot_model(t, rates):
    plt.plot(t, rates)
    plt.xlabel('Time (t)')
    plt.ylabel('Rates')
    plt.title('Vasicek Model')
    plt.show()
```

```python
plot_model(*vasicek_model(1.3, 0.9, 5, 100))
```
